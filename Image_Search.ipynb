```python
from DeepImageSearch import Load_Data,Search_Setup
```

```python
# Load images from a folder
image_list = Load_Data().from_folder(['drive/MyDrive/archive'])
```

```python
print("Total Image Count:",len(image_list))
print("Samples:")
print(image_list[:10])
```

```python
# Set up the search engine
st = Search_Setup(image_list=image_list,model_name='vgg19',pretrained=True,image_count= 100)
```

```python
# Index the images
st.run_index()
```

```python
# Get metadata
metadata = st.get_image_metadata_file()
metadata
```

```python
# Add new images to the index
st.add_images_to_index(image_list[1001:1010])
```

```python
# Update metadata
metadata = st.get_image_metadata_file()
metadata
```

```python
# Get similar images
st.get_similar_images(image_path=image_list[11],number_of_images=10)
```

```python
# Plot similar images
st.plot_similar_images(image_path = image_list[307],number_of_images=16)
```
